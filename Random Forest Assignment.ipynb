```python
q1=pd.read_csv('Company_data.csv')
```

```python
q1.head()
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
q1["ShelveLoc"]=le.fit_transform(q1["ShelveLoc"])
q1["Urban"]=le.fit_transform(q1["Urban"])
q1["US"]=le.fit_transform(q1["US"])
```

```python
q1=q1.assign(Sale=pd.cut(q1['Sales'], 
                               bins=[ 0, 4, 9,15], 
                               labels=['Low', 'Medium', 'High']))
```

```python
q1.head(10)
```

```python
q1['Sale']=le.fit_transform(q1['Sale'])
```

```python
q1 = q1['Sale'].fillna('').apply(str)
```

```python
q1.info()
```

```python
q1.drop('Sales',axis=1,inplace=True)
```

```python
x=q1.iloc[:,0:10]
y=q1.iloc[:,-1]
```

```python
x.head()
```

```python
y.head(10)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
base_model=LogisticRegression()
```

```python
base_model1=DecisionTreeClassifier()
```

```python
rf_model=RandomForestClassifier(n_estimators=50,max_depth=4,max_features=3)  
```

```python
rf_model.fit(x_train,y_train)
```

```python
y_pred=rf_model.predict(x_train)
```

```python
confusion_matrix(y_train,y_pred)
```

```python
print(classification_report(y_train,y_pred))
```

```python
q2=pd.read_csv('Fraud_check.csv')
```

```python
q2
```

```python
q2.info()
```

```python
DF=pd.Series(q2['Taxable.Income'])
su=[]
for i in DF:
    if i<=30000:
        su.append("risky")
    else:
        su.append('good')
print(su)
```

```python
DF2=pd.DataFrame(su)
DF2=pd.concat([DF2,q2],axis=1)
DF2=DF2.rename({0:'o/p'},axis=1)
```

```python
DF2
```

```python
labelencoder=LabelEncoder()
DF2.iloc[:,0]=labelencoder.fit_transform(DF2.iloc[:,0])
DF2.iloc[:,1]=labelencoder.fit_transform(DF2.iloc[:,1])
DF2.iloc[:,2]=labelencoder.fit_transform(DF2.iloc[:,2])
DF2.iloc[:,-1]=labelencoder.fit_transform(DF2.iloc[:,-1])
```

```python
DF2
```

```python
DF2=DF2.values
```

```python
x=DF2[:,1:7]
y=DF2[:,0]
```

```python
num_trees = 100
max_features = 4
kfold = KFold(n_splits=10)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```
